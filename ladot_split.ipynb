```python
import pandas as pd
import numpy as np
import datetime
```

```python
filename='LADOT_Traffic_Counts_Summary.csv'
years=['2010','2011','2012','2013','2014','2015','2016']
df=pd.read_csv('LADOT_Traffic_Counts_Summary.csv',sep=',',dtype=str)
```

```python
df
```

```python
dffilter=df[(df["Type"]=='AUT') | (df["Type"]=='AUTO')]
```

```python
dffilter
```

```python
dffilter=dffilter.iloc[pd.to_datetime(dffilter["Count Date"],format='%B %d %Y',errors='coerce').values.argsort()]
```

```python
dffilter
```

```python
for year in years:
    dftmp=dffilter[dffilter['Count Date'].str.contains(year,na=False)]
    dftmp.to_csv(filename[:-4]+year+filename[-4:],index=False)
    print(dftmp)
```
