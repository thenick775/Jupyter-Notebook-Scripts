```python
import pandas as pd
import numpy as np
```

```python
filename="suomi_npp_omps_l2ovp_nmso2_v01_guangzhou_655.txt"
year="2018"
resfile=filename[:-4]+"_"+str(year)+filename[-4:]

#file string spacing formatting
myl1=lambda x:  ' '*(13-len(x))+x
myl2=lambda x:  ' '*(5-len(x))+x
myl3=lambda x:  ' '*(4-len(x))+x
myl4=lambda x:  ' '*(10-len(x))+x
myl5=lambda x:  ' '*(7-len(x))+x
myl6=lambda x:  ' '*(14-len(x))+x
```

```python
df=pd.read_csv(filename,delimiter="\s+",skiprows=30,dtype=str, names=["Datetime","MJD2000","Year","DOY","sec. (UT)","Orbit","CTP","Lat.","Lon.","Dist.","SZA","SO2_PBL","SO2_TRL","SO2_TRM","SO2_TRU","SO2_STL","OMTO3_O3","Cld. F.","OMTO3_Ter_Pre","OMTO3_Ref","AI"])
```

```python
df
```

```python
header=''
count=0
countbool=False
with open(filename,'r') as p:
    for line in p.readlines():
        header+=line
        if "Read format (FORTRAN/IDL):" in line:
            countbool=True
            #header+='\n'
            #break
        if countbool:
            count+=1
        if count==3:
            break
```

```python
header
```

```python
dffilter=df[df["Year"]==year].copy()
```

```python
dffilter
```

```python
#format each string cell
dffilter["MJD2000"]=dffilter["MJD2000"].apply(myl1)
dffilter[["Year","CTP"]]=dffilter[["Year","CTP"]].applymap(myl2)
dffilter["DOY"]=dffilter["DOY"].apply(myl3)
dffilter[["sec. (UT)","SO2_PBL","SO2_TRL","SO2_TRM","SO2_TRU","SO2_STL","OMTO3_O3","Cld. F.","OMTO3_Ref","AI"]]=dffilter[["sec. (UT)","SO2_PBL","SO2_TRL","SO2_TRM","SO2_TRU","SO2_STL","OMTO3_O3","Cld. F.","OMTO3_Ref","AI"]].applymap(myl4)
dffilter[["Orbit","Lat.","Lon.","Dist.","SZA"]]=dffilter[["Orbit","Lat.","Lon.","Dist.","SZA"]].applymap(myl5)
dffilter["OMTO3_Ter_Pre"]=dffilter["OMTO3_Ter_Pre"].apply(myl6)
```

```python
with open(resfile,"w") as f:
    f.write("%s%s"%(header,dffilter.to_csv(index=False,header=False,sep=" ").replace('"', '')))
```
