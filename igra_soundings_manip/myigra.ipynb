```python
df2=df.loc[df[1] == '#USM00070026']#separate the rows that contain <- as the first entruy
```

```python
df2
```

```python
df2=df2.reset_index(drop=True)#reset the index so we can align both dataframes
```

```python
df2=df2.reindex(df2.index.repeat(df2[7]))
```

```python
df2=df2.reset_index(drop=True)
```

```python
df2
```

```python
df_new=df2
```

```python
#now for below, get all rows that do not start with this designation
```

```python
df3=df.loc[df[1]!='#USM00070026']#assign it to a new dataframe
```

```python
df3
```

```python
df3=df3.reset_index(drop=True)#reset the index for the merge
```

```python
df3
```

```python
result=pd.concat([df_new, df3], axis=1)
```
