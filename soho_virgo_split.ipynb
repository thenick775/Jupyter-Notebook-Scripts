```python
import pandas as pd
import numpy as np
from datetime import date
```

```python
df=pd.read_csv("VIRGO_TSI_hourly.dat",sep="\s+",skiprows=37,header=None,dtype=str)
```

```python
myl1=lambda x: ' '*(10-len(str(x)))+str(x)
downlimasint=20180101
uplimasint=20190101
finfilename='VIRGO_TSI_hourly_2018.dat'
```

```python
headerstr=''
i=0
with open("VIRGO_TSI_hourly.dat",'r') as f:
    for line in f.readlines():
        #print(i)
        headerstr+=line
        i+=1
        if i>37:
            break;
```

```python
headerstr
```

```python
df
```

```python
dfyearspl2=df[(df[0].astype(int)>=downlimasint) & (df[0].astype(int)<uplimasint)]
dfyearspl2=dfyearspl2.reset_index(drop=True)
```

```python
dfyearspl2
```

```python
dfyearspl2[[2,3,4,5]]=dfyearspl2[[2,3,4,5]].applymap(myl1)
```

```python
strdat=dfyearspl2.to_csv(sep=' ',header=False,index=False).replace('"','')
```

```python
strtot='%s%s'%(headerstr,strdat)
```

```python
strtot
```

```python
with open(finfilename,'w') as p:
    p.write(strtot)
```
