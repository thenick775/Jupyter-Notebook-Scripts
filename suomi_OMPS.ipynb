```python
import pandas as pd
import numpy as np
```

```python
filename="suomi_npp_omps_l2ovp_nmso2_v01_cabauw_818.txt"
year="2018"
resfile=filename[:-4]+"_"+str(year)+filename[-4:]
isso2=True#determiness what column names amd spacing to use

colnames=["Datetime","MJD2000","Year","DOY","sec. (UT)","Orbit","CTP","Lat.","Lon.","Dist.","SZA","SO2_PBL","SO2_TRL","SO2_TRM","SO2_TRU","SO2_STL","OMTO3_O3","Cld. F.","OMTO3_Ter_Pre","OMTO3_Ref","AI"] if isso2 else ["Datetime", "MJD2000", "Year", "DOY", "sec. (UT)", "Orbit", "CTP", "Lat.", "Lon.", "Dist.", "SZA", "Ozone", "O3blwCld", "Surf. P.", "Cld. P.", "Cld. F.", "Ref.", "AI", "SOI"]
#above uses ternary operator based on isso2 to decide the column names
#file string spacing formatting, applied to each cell in the selected column (or applymap if more than 1 column)
myl1=lambda x:  ' '*(13-len(x))+x
myl2=lambda x:  ' '*(5-len(x))+x
myl3=lambda x:  ' '*(4-len(x))+x
myl4=lambda x:  ' '*(10-len(x))+x
myl5=lambda x:  ' '*(7-len(x))+x
myl6=lambda x:  ' '*(14-len(x))+x
myl7=lambda x:  ' '*(8-len(x))+x
myl8=lambda x:  ' '*(9-len(x))+x
myl9=lambda x:  ' '*(6-len(x))+x
```

```python
df=pd.read_csv(filename,delimiter="\s+",skiprows=30,dtype=str, names=colnames)
```

```python
df
```

```python
header=''
count=0
countbool=False
with open(filename,'r') as p:
    for line in p.readlines():
        header+=line
        if "Read format (FORTRAN/IDL):" in line:
            countbool=True
            #header+='\n'
            #break
        if countbool:
            count+=1
        if count==3:
            break
```

```python
header
```

```python
dffilter=df[df["Year"]==year].copy()
```

```python
dffilter
```

```python
#format each string cell
if isso2:#format for so2 files
    dffilter["MJD2000"]=dffilter["MJD2000"].apply(myl1)
    dffilter[["Year","CTP"]]=dffilter[["Year","CTP"]].applymap(myl2)
    dffilter["DOY"]=dffilter["DOY"].apply(myl3)
    dffilter[["sec. (UT)","SO2_PBL","SO2_TRL","SO2_TRM","SO2_TRU","SO2_STL","OMTO3_O3","Cld. F.","OMTO3_Ref","AI"]]=dffilter[["sec. (UT)","SO2_PBL","SO2_TRL","SO2_TRM","SO2_TRU","SO2_STL","OMTO3_O3","Cld. F.","OMTO3_Ref","AI"]].applymap(myl4)
    dffilter[["Orbit","Lat.","Lon.","Dist.","SZA"]]=dffilter[["Orbit","Lat.","Lon.","Dist.","SZA"]].applymap(myl5)
    dffilter["OMTO3_Ter_Pre"]=dffilter["OMTO3_Ter_Pre"].apply(myl6)
else:#apply slightly different formatting here for o3 files
    dffilter["MJD2000"]=dffilter["MJD2000"].apply(myl1)
    dffilter[["Year","CTP"]]=dffilter[["Year","CTP"]].applymap(myl2)
    dffilter["DOY"]=dffilter["DOY"].apply(myl3)
    dffilter[["sec. (UT)","Cld. F."]]=dffilter[["sec. (UT)","Cld. F."]].applymap(myl4)
    dffilter[["Orbit","Lat.","Lon.","SOI"]]=dffilter[["Orbit","Lat.","Lon.","SOI"]].applymap(myl5)
    dffilter[["SZA","Ozone"]]=dffilter[["SZA","Ozone"]].applymap(myl7)
    dffilter[["O3blwCld","Surf. P.","Cld. P.","Ref.","AI"]]=dffilter[["O3blwCld","Surf. P.","Cld. P.","Ref.","AI"]].applymap(myl8)
    dffilter["Dist."]=dffilter["Dist."].apply(myl9)
```

```python
with open(resfile,"w") as f:
    f.write("%s%s"%(header,dffilter.to_csv(index=False,header=False,sep=" ").replace('"', '')))
```
