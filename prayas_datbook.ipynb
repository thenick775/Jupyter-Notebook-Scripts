```python
import pandas as pd
import numpy as np
```

```python
loc_to_state_dict={}
```

```python
df=pd.read_csv('ESMI location information.csv')
dfdat1=pd.read_csv('ESMI minute-wise voltage data 2018 Jan-May.csv',low_memory=False)
dfdat2=pd.read_csv('ESMI minute-wise voltage data 2018 June-Aug.csv',low_memory=False)
dfdat3=pd.read_csv('ESMI minute-wise voltage data 2018 Sep-Dec.csv',low_memory=False)
```

```python
df
```

```python
statelist=df["State"].unique()
print(statelist)
```

```python
for state in statelist:
    loc_to_state_dict.update({state:[]})
print(loc_to_state_dict)
```

```python
for index,row in df.iterrows():
    print(str(index)+' '+row[0]+' '+row[4])
    
    if not row[0] in loc_to_state_dict[row[4]]:
        loc_to_state_dict.update({row[4]:loc_to_state_dict[row[4]]+[row[0]]})

'''for index,row in dfdat2.iterrows():
    print(index)
    print(row[4])
for index,row in dfdat3.iterrows():
    print(index)
    print(row[4])'''
print(loc_to_state_dict)
```

```python
for state in loc_to_state_dict:
    fin=pd.DataFrame()
        
    for location in loc_to_state_dict[state]:
        dftmp1=dfdat1.loc[(dfdat1['Location name']==location)]
        dftmp2=dfdat2.loc[(dfdat1['Location name']==location)]
        dftmp3=dfdat3.loc[(dfdat1['Location name']==location)]
        fin=fin.append(dftmp1,ignore_index=True)
        fin=fin.append(dftmp2,ignore_index=True)
        fin=fin.append(dftmp3,ignore_index=True)
        #print(dftmp1)
        #print(dftmp2)
        #print(dftmp3)
    fin.to_csv('%s%s%s'%('ESMI minute-wise voltage data 2018 ',state if not '/' in state else state.replace('/','-'),'.csv'),index=False)
    #break;
```
