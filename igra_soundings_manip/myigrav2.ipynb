```python
import pandas as pd
import numpy as np
```

```python
#splits the header and data into separate files with no editing
filename='USM00070026-data.txt'#origional file name
datname='testdat.txt' #temporary file for data
headname='testhead.txt' #temporary file for header, these can be removed through os.remove() if we choose
res_fname="finigratest.txt"

with open(filename,'r') as f:#origional file
    with open(datname,'w') as testdat:#data file
        with open(headname,'w') as testhead:#header file
            for line in f.readlines():#process by individual line, as it could be header or data
                if "#USM" in line:#write to either the header or data file without modification as files
                    testhead.write(line)#may have to modify this per source file, as I am assumign every header line will contain this
                else:
                    testdat.write(line)#else write to the data file
```
