```python
import numpy as np
import pandas as pd
```

```python
name='2018.l3i'
N=3#for header
col_names=['year','month','day','doy','hour','minute','status','accum.time','int_p4','int_p8','int_p25','int_p41','sys_p4','sys_p8','sys_p25','sys_p41','stat_p4','stat_p8','stat_p25','stat_p41','int_h4','int_h8','int_h25','int_h41','sys_h4','sys_h8','sys_h25','sys_h41','stat_h4','stat_h8','stat_h25','stat_h41']

pname=name[:-4]+'_proton'+name[4:]
print(pname)
hname=name[:-4]+'_helium'+name[4:]
print(hname)

df=pd.read_csv(name,delim_whitespace=True,skiprows=3,header=None,dtype=str)
df.columns=col_names
```

```python
def head_manip(header,bll):#true for proton, false for helium
    #print(header[0])
    tmp_headpr=header[0]#no need to copy with strings
    tmp_headhe=header[0]
    #print(dfpr.columns)
    #print('from proton')
    for col in dfpr.columns[8:32]:
        #print(col)
        tmp_headhe=tmp_headhe.replace(col,'',1)
    tmp_headhe=tmp_headhe.replace('                 ',' ',1)#correct helium for leading spaces after date cols
    #print('from helium')
    for col in dfhefin.columns[8:32]:
        #print(col)
        tmp_headpr=tmp_headpr.replace(col,'',1)
    #print(header[0])
    if bll:
        #print('from pro')
        header[0]=tmp_headpr.strip()+'\n'
    else:
        #print('from he')
        header[0]=tmp_headhe.strip()+'\n'
    #print(''.join(header))
    return ''.join(header)
```

```python
with open(name) as myfile:
    head = [next(myfile) for x in range(N)]

#header_str=''.join(head)
dfpr_header_str=head_manip(head.copy(),True)
dfhe_header_str=head_manip(head.copy(),False)
print('orig head=\n'+''.join(head)+'\npr head=\n'+''.join(dfpr_header_str)+'\nhe head=\n'+''.join(dfhe_header_str))
#dfprf_str=dfpr.to_csv(index=False,header=False,sep=' ')
#dfhef_str=dfhefin.to_csv(index=False,header=False,sep=' ')
dfprf_str=data_spacing_manip(dfpr.copy())
dfhef_str=data_spacing_manip(dfhefin.copy())
with open(pname, 'w') as pf, open(hname, 'w') as hf:
    pf.write('%s%s'%(dfpr_header_str,dfprf_str))
    hf.write('%s%s'%(dfhe_header_str,dfhef_str))
print('---finished---')
```
